```python
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
```

```python
result_list = []

for col in map(str, range(9)):  # Loop over column names as strings from '0' to '8'
    gene = TL_marker[col].tolist()  # Convert the column to a list
    result = sctl.ai.prioritize_genes(
        gene_list=gene,
        context="cells isolated from mouse lung tissue",
        api_token=OPENAI_API_KEY,
        n=20,
        platform="openai",
        llm_model="o4-mini")
    result_list.append(result)
print(result_list)
```
